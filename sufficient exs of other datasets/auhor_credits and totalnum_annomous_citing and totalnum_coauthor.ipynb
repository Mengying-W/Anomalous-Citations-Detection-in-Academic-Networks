# author credit

```python
def getCredit_2(f1,f2):
    authors_List = []
    authors_num = 0
    total_vex = 0
    for line in f1:
        line = line.strip()
        sen = eval(line)
        total_vex = total_vex +1
        authors = sen['authors'] #LIST
        for a in authors:
            if a not in authors_List:
                authors_List.append(a)
                
                if total_vex <= 13:           
                    f2.write(str(1)+'\n')
                      
                elif total_vex > 13 and total_vex <= 26:
                    f2.write(str(0)+'\n')
                       
                elif total_vex >26 and total_vex <= 181:
                    f2.write(str(1)+'\n')  
                    
                elif total_vex >181 and total_vex <= 274:
                    f2.write(str(0)+'\n') 
                    
				
                authors_num = authors_num + 1
                                  
    print("authors_num="+str(authors_num))
    print("total_vex="+str(total_vex))
```

# coauthor

```python
Wtemp = np.zeros((1207,645))     #作者-文章矩阵
f = open('/Users/carina/Downloads/刘嘉莹/异常引用补充实验/sourcecode/ANCI_code/dataset/coauthor_refer/processed data/Authors_paper_matrix.txt')
lines = f.readlines()
    
Wtemp_row = 0
for line in lines:
    list1 = line.strip().split(' ')
        #print(len(list1))
    Wtemp[Wtemp_row:] = list1[0:645]
    Wtemp_row += 1
```

```python
sum(map(sum,Wtemp)) #引用链接数量
```

```python
f = open('/Users/carina/Downloads/刘嘉莹/异常引用补充实验/sourcecode/ANCI_code/dataset/coauthor_refer/processed data/Author_author_matrix.txt')
A = np.ones((1207,1207))
lines = f.readlines()
A_row = 0
for line in lines:
    list = line.strip().split(' ')
    print(len(list))
    A[A_row:] = list[0:1207]
    A_row += 1

f.close()
print ("-----------Author_author_matrix.txt prepared-------------------")
```

```python
s = 0
for i in range(len(A)):
    s += A[i][i]
    i += 1
print(s)
```

```python
h = (sum(map(sum,A)) - s)/2
h
```
